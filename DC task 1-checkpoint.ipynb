```python
import pandas as pd
import numpy as np 
from sklearn.tree import DecisionTreeRegressor
```

```python
data_1=pd.read_excel(r"C:\Users\USER\Desktop\Activity  2024-25\Data Mining\Students Buys Computer.xlsx")
data_1
```

```python
data_1.columns
```

```python
cleanup_nums = {'Age':{"<30": 1, "31-40": 2,">40":3}, 'Income':{"High": 1, "Medium": 2, "Low": 3},'Students':{"Yes":1,"No":2}}
```

```python
cleanup_nums = {'Age':{"<30": 1, "31-40": 2,">40":3}, 'Income':{"High": 1, "Medium": 2, "Low": 3},'Students':{"Yes":1,"No":2},'Credit Salary ':{"Fair":1,"Excellent":2}, 'Buys Computer':{"Yes":1,"No":2}}
```

```python
df = data_1.replace(cleanup_nums)
#data1.to_csv("Data_3_var.csv")
df.head()
```

```python
regressor = DecisionTreeRegressor(random_state = 0,max_depth=2)
x = df.drop(['Buys Computer'],axis=1)
y = df['Buys Computer']
#Fit the regressor object to the dataset.
regressor.fit(x,y)
```

```python
y_pred = regressor.predict(x)
y_pred
```

```python
print("Coefficient of determination R^2 <-- on train set: {}".format(regressor.score(x, y)))
```

```python
from sklearn import tree 
```

```python
tree.plot_tree(regressor,feature_names=['Age', 'Income', 'Students', 'Credit Salary '])
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
from sklearn import tree
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
clf = DecisionTreeClassifier()

# Train the classifier
clf.fit(x_train, y_train)

# Make predictions on the test set
y_pred = clf.predict(x_test)

# Evaluate the model's performance
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")

# Detailed classification report
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=data_1['Buys Computer'].unique()))
# Confusion matrix
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

# Visualizing the Decision Tree
plt.figure(figsize=(12,8))
tree.plot_tree(clf, feature_names=['Age', 'Income', 'Students', 'Credit Salary '], class_names=['Buys Computer'], filled=True)
plt.show()
```
