# pearson's correlation

```python
x = ['Averag marks scored before pandemic in traditional classroom']
y = ['Study time (Hours)' ]
corr_coeff, p_value = spearmanr(x, y)

print("Spearman's rank correlation coefficient:", corr_coeff)
print("P-value:", p_value)
```

# kruskal walis

```python
groups = [data[data['Level of Education'] == group][ 'Study time (Hours)'] for group in data['Level of Education'].unique()]

# Performing Kruskal-Wallis test
h_statistic, p_value = stats.kruskal(*groups)

# Printing the results
print("H-Statistic:", h_statistic)
print("P-Value:", p_value)

# Checking significance
alpha = 0.05
if p_value < alpha:
    print("Reject null hypothesis: There is significant evidence that at least one group median is different.")
else:
    print("Fail to reject null hypothesis: There is no significant evidence that the group medians are different.")
```

# chi Square

```python
contingency_table=pd.crosstab(data['Level of Education'],data['Device type used to attend classes'])
contingency_table
```

```python
chi2_stat,p_value,dof,expected= chi2_contingency(contingency_table)
print(f"Chi2 Stat: {chi2_stat}")
print(f"P-value: {p_value}")
print(f"Degrees of freedom: {dof}")
print(f"Expected frequencies table:")
print(expected)
```
